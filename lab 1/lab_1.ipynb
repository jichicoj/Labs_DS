### PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline

pca_pipe = make_pipeline(StandardScaler(),PCA()) #Se escalan los datos y luego se le aplica PCA
pca_pipe.fit(data[n_numerical_cols])
#Se extrae el modelo del pipeline
modelo_pca = pca_pipe.named_steps['pca']

#Se convierte el arreglo en data frame
pd.DataFrame(
    data = modelo_pca.components_,
    columns = n_numerical_cols,
    index = ['PC1','PC2','PC3','PC5','PC6','PC7','PC8', 'PC9', 'PC10', 'PC11']
)
```

#### Porcentaje de la varianza acumulada

```python
prop_varianza_acum = modelo_pca.explained_variance_ratio_.cumsum()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))
ax.plot(
    np.arange(len(n_numerical_cols)) + 1,
    prop_varianza_acum,
    marker = 'o'
)

for x, y in zip(np.arange(len(n_numerical_cols)) + 1, prop_varianza_acum):
    label = round(y, 2)
    ax.annotate(
        label,
        (x,y),
        textcoords="offset points",
        xytext=(0,10),
        ha='center'
    )
    
ax.set_ylim(0, 1.1)
ax.set_xticks(np.arange(modelo_pca.n_components_) + 1)
ax.set_title('Porcentaje de varianza explicada acumulada')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza acumulada');
```

Veamos que en este caso, no parece ser que ninguna de las componentes esté fuertemente compuesta por una o pocas variables, sino que parece que todas son una mezcla algo aleatoria de las variables que tenemos. Sin mebargo, veamos que la 1 componente principal está compuesta fuertemente por la edad, si la persona fuma y la cantidad de ETS que la persona ha tenido. Esto tiene sentido ya que médicamente estos son de los mayores causantes de cáncer en la mujeres, y lo de las ETS se peude ver más claramente en la componente principal 2. Esto también nos indica que factores como el momento en el que se tuvo la primera relación sexual, o si la persona tuvo o no tuvo hijos, no afecta de manera significativa la posibilidad de una mujer de desarrolalr cáncer.
