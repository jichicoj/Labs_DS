### Variables description

#### Quantitative variables
- Age -> age of the patient (discrete)
- Number of sexual partners -> number of sexual partners that the patient has ever had (discrete)
- First sexual intercourse -> age of the patient when she first had sexual intercourse (discrete)
- Num of pregnancies -> number of pregnancies that the patient has had (discrete)
- Smokes (years) -> if the patient smokes, the number of years that she has been doing it (continuous)
- Smokes (packs/year) -> if the patient smokes, the number of packs she smokes per year (continuous)
- Hormonal Contraceptives (years) -> if the patient uses hormonal contraceptives, the number of years she has been using it (continuous)
- IUD (years) -> if the patient has a IUD, the number of years that she has it (continuous)
- STDs (number) -> if the patient has had any STDs, the number of STDs that she has ever had (discrete)
- STDs: Number of diagnosis -> number of times the patient has been diagnosed with a STD (discrete)
- STDs: Time since first diagnosis -> time that has gone by since the patient was first diagnosed with a STD (discrete)
- STDs: Time since last diagnosis -> time that has gone by since the patient was last diagnosed with a STD (discrete)

#### Qualitative variables
- Smokes -> if the patient smokes or not
- Hormonal Contraceptives -> if the patient uses hormonal contraceptives
- IUD -> if the patient has a IUD
- STDs -> if the patient has ever had STDs
- STDs:condylomatosis -> if the patient has had condylomatosis
- STDs:cervical condylomatosis -> if the patient has had cervical condylomatosis
- STDs:vaginal condylomatosis -> if the patient has had vaginal condylomatosis
- STDs:vulvo-perineal condylomatosis -> if the patient has had vulvo-perineal condylomatosis
- STDs:syphilis -> if the patient has had syphilis
- STDs:pelvic inflammatory disease -> if the patient has had pelvic inflammatory disease
- STDs:genital herpes -> if the patient has had genital herpes
- STDs:molluscum contagiosum -> if the patient has had molluscum contagiosum
- STDs:AIDS -> if the patient has AIDS
- STDs:HIV -> if the patient has HIV
- STDs:Hepatitis B -> if the patient has or has had hepatitis B
- STDs:HPV -> if the patient has of has had HPV
- Dx:Cancer -> if the patient has a cancer diagnostic
- Dx:CIN -> if the patient has a CIN diagnostic
- Dx:HPV -> if the patient has a HPV diagnostic
- Dx -> if the patient has any diagnostic
- Hinselmann -> if the patient had a colonoscopy for her diagnostic
- Schiller -> if the doctors did a Schiller test for the patients diagnostic
- Citology -> if the doctors did a citology for the patients diagnostic
- Biopsy -> if the doctors did a biopsy for the patients diagnostic

```python
numerical_cols = ['Age', 'First sexual intercourse', 'Number of sexual partners', 'Num of pregnancies', 'Smokes (years)', 'Smokes (packs/year)', 'Hormonal Contraceptives (years)', 'IUD (years)', 'STDs (number)', 'STDs: Number of diagnosis', 'STDs: Time since first diagnosis', 'STDs: Time since last diagnosis']
categorical_cols = list(set(list(data.keys())) - set(numerical_cols))
```

```python
for col in numerical_cols:
    print(data[col].value_counts())
```

```python
for col in numerical_cols:
    data[col] = data[col].replace('?', np.nan)
```

```python
for col in categorical_cols:
    print(data[col].value_counts())
```

```python
for col in categorical_cols:
    data[col] = data[col].replace('?', 2)
```

```python
data = data.apply(pd.to_numeric, errors='coerce')
```

```python
num_vars = len(numerical_cols)
num_cols = 3  # Number of columns in the plot grid
num_rows = (num_vars + num_cols - 1) // num_cols  # Number of rows needed

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 5))
axes = axes.flatten()

for i, col in enumerate(numerical_cols):
    sns.histplot(data[col], kde=True, bins=30, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
num_vars = len(numerical_cols)
num_cols = 3  # Number of columns in the plot grid
num_rows = (num_vars + num_cols - 1) // num_cols  # Number of rows needed

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 5))
axes = axes.flatten()

for i, col in enumerate(numerical_cols):
    sns.boxplot(y=col, data=data, ax=axes[i])
    axes[i].set_title(f'Boxplot of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel(col)

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
num_vars = len(categorical_cols)
num_cols = 3  # Number of columns in the plot grid
num_rows = (num_vars + num_cols - 1) // num_cols  # Number of rows needed

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 5))
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
    sns.countplot(x=col, data=data, ax=axes[i])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
correlation_matrix = data[numerical_cols].corr()

# Plot the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```
